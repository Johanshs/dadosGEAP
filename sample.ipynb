```python
import pandas as pd

df = pd.read_csv('GEAP.csv', sep=';')
df.columns = df.columns.str.strip()

# usando a planilha reduzida
df_teste = df.head(13)
display(df)
# print(f"Número de linhas: {df_teste.shape[0]}, Número de colunas: {df_teste.shape[1]}")
# contagem_titulares = df_teste[df_teste['TipoVinculo'] == 'TITULAR'].shape[0]
# contagem_dependentes = df_teste[df_teste['TipoVinculo'] == 'FILHO(A)'].shape[0]
# print(f"Número de titulares: {contagem_titulares}")
# print(f"Número de dependentes: {contagem_dependentes}")

class Pessoas:
    def __init__(self, cpf, nome, conveniada, data_nascimento, seqcliente, tipo_vinculo, plano) -> None:
        self.cpf = cpf
        self.nome = nome
        self.conveniada = conveniada
        self.data_nascimento = data_nascimento
        self.seqcliente = seqcliente
        self.tipo_vinculo = tipo_vinculo
        self.plano = plano
    

class Servidores:
    def __init__(self, pessoa:Pessoas, cod_matricula, nroinscricao) -> None:
        self.pessoa = pessoa
        self.cod_matricula = cod_matricula
        self.nroinscricao = nroinscricao
        self.dependentes = []
        
    def add_dependente(self, dependente):
        # Verificar se o dependente já está vinculado
        if dependente not in self.dependentes:
            self.dependentes.append(dependente)
        else:
            print(f"Dependente {dependente.pessoa.nome} já está vinculado a este titular.")

    def rem_dependente(self, dependente):
        if dependente in self.dependentes:
            self.dependentes.remove(dependente)
        else:
            print(f"Dependente {dependente.pessoa.nome} não está vinculado a este titular.")

        
class Dependentes:
    def __init__(self, pessoa:Pessoas,nroinscricao ) -> None:
        self.pessoa = pessoa
        self.nroinscricao = nroinscricao
        
import pandas as pd

# Listas
titulares: list[Servidores] = []
dependentes: list[Dependentes] = []

# Criar um dicionário para mapear servidores pela matrícula
servidor_por_matricula = {}

for i, row in df_teste.iterrows():
    # Criar a pessoa (tanto para titulares quanto dependentes)
    pessoa = Pessoas(
        cpf=row['CPF'], 
        nome=row['NmeCliente'], 
        conveniada=row['Conveniada'], 
        data_nascimento=row['DtaInscricao'], 
        seqcliente=row['SeqCliente'], 
        tipo_vinculo=row['TipoVinculo'], 
        plano=row['Plano']
    )
    
    if row['TipoVinculo'] == "TITULAR":
        servidor = Servidores(pessoa, cod_matricula=row['CodMatricula'], nroinscricao=row['NroInscricao'])
        titulares.append(servidor)
        servidor_por_matricula[row['CodMatricula']] = servidor  # Mapeia o servidor pela matrícula
    else:
        dependente = Dependentes(pessoa, nroinscricao=row['NroInscricao'])
        # Adiciona o dependente à lista do servidor correspondente
        cod_matricula = row['CodMatricula']
        if cod_matricula in servidor_por_matricula:
            servidor_por_matricula[cod_matricula].add_dependente(dependente)  # Adiciona o dependente ao servidor
        dependentes.append(dependente)  # Opcional: manter a lista de dependentes, caso precise dela


# Criar uma lista para armazenar os dados para o DataFrame
dados = []

# Preencher a lista com dados dos servidores e dependentes
for servidor in titulares:
    # Adiciona o servidor (titular)
    dados.append({
        'Nome': servidor.pessoa.nome,
        'CPF': servidor.pessoa.cpf,
        'Matrícula': servidor.cod_matricula,
        'NroInscricao': servidor.nroinscricao,
        'Vinculo': servidor.pessoa.tipo_vinculo
    })
    
    # Adiciona os dependentes do servidor
    for dependente in servidor.dependentes:
        dados.append({
            'Nome': dependente.pessoa.nome,
            'CPF': dependente.pessoa.cpf,
            'Matrícula': servidor.cod_matricula,  # Matrícula do titular
            'NroInscricao': dependente.nroinscricao,
            'Vinculo': dependente.pessoa.tipo_vinculo
        })

# Criação do DataFrame
df_refeito = pd.DataFrame(dados)

# Exibir o DataFrame
display(df_refeito)
```
