```python
# Uso das classes sem array de dependentes na classe servidor
import pandas as pd

df = pd.read_csv('GEAP.csv', sep=';')
df.columns = df.columns.str.strip()

# usando a planilha reduzida
df_teste = df.head(13)
display(df_teste)

class Pessoas:
    def __init__(self, cpf, nome, conveniada, data_nascimento, seqcliente, tipo_vinculo, plano) -> None:
        self.cpf = cpf
        self.nome = nome
        self.conveniada = conveniada
        self.data_nascimento = data_nascimento
        self.seqcliente = seqcliente
        self.tipo_vinculo = tipo_vinculo
        self.plano = plano
    

class Servidores:
    def __init__(self, pessoa:Pessoas, cod_matricula, nroinscricao) -> None:
        self.pessoa = pessoa
        self.cod_matricula = cod_matricula
        self.nroinscricao = nroinscricao
        

        
class Dependentes:
    def __init__(self, pessoa:Pessoas,nroinscricao ) -> None:
        self.pessoa = pessoa
        self.nroinscricao = nroinscricao
        
# Listas
titulares: list[Servidores] = []
dependentes: list[Dependentes] = []

for i, row in df_teste.iterrows():
    # Criar a pessoa (tanto para titulares quanto dependentes)
    pessoa = Pessoas(
        cpf=row['CPF'], 
        nome=row['NmeCliente'], 
        conveniada=row['Conveniada'], 
        data_nascimento=row['DtaInscricao'], 
        seqcliente=row['SeqCliente'], 
        tipo_vinculo=row['TipoVinculo'], 
        plano=row['Plano']
    )
    
    if row['TipoVinculo'] == "TITULAR":
        servidor = Servidores(pessoa, cod_matricula=row['CodMatricula'], nroinscricao=row['NroInscricao'])
        titulares.append(servidor)
    else:
        dependente = Dependentes(pessoa, nroinscricao=row['NroInscricao'])
        dependentes.append(dependente)
    
    
print("="*20, end="\n\n")

for titular in titulares:
    print(f"Titular: {titular.pessoa.nome}")
    dependentes_encontrados = []
    
    # Procurar dependentes com o mesmo número de inscrição
    for dependente in dependentes:
        if dependente.nroinscricao == titular.nroinscricao:
            dependentes_encontrados.append(dependente.pessoa.nome)
    
    if dependentes_encontrados:
        print(f"Dependentes : {', '.join(dependentes_encontrados)} \n")
    else:
        print("Nenhum dependente encontrado para este titular.")
```

```python
# Transformando os dados e refazendo o dataframe
import pandas as pd

df = pd.read_csv('GEAP.csv', sep=';')
df.columns = df.columns.str.strip()

df_teste = df.head(21)
display(df_teste)

class Pessoas:
    def __init__(self, cpf, nome, conveniada, data_nascimento, seqcliente, tipo_vinculo, plano) -> None:
        self.cpf = cpf
        self.nome = nome
        self.conveniada = conveniada
        self.data_nascimento = data_nascimento
        self.seqcliente = seqcliente
        self.tipo_vinculo = tipo_vinculo
        self.plano = plano
    

class Servidores:
    def __init__(self, pessoa:Pessoas, cod_matricula, nroinscricao) -> None:
        self.pessoa = pessoa
        self.cod_matricula = cod_matricula
        self.nroinscricao = nroinscricao
        self.dependentes = []
        
    def add_dependente(self, dependente):
        # Verificar se o dependente já está vinculado
        if dependente not in self.dependentes:
            self.dependentes.append(dependente)
        else:
            print(f"Dependente {dependente.pessoa.nome} já está vinculado a este titular.")

    def rem_dependente(self, dependente):
        if dependente in self.dependentes:
            self.dependentes.remove(dependente)
        else:
            print(f"Dependente {dependente.pessoa.nome} não está mais vinculado a este titular.")

        
class Dependentes:
    def __init__(self, pessoa:Pessoas,nroinscricao ) -> None:
        self.pessoa = pessoa
        self.nroinscricao = nroinscricao
        
import pandas as pd

# Listas
titulares: list[Servidores] = []
dependentes: list[Dependentes] = []

# Criar um dicionário para mapear servidores pela matrícula
servidor_por_matricula = {}

for i, row in df_teste.iterrows():
    # Criar a pessoa (tanto para titulares quanto dependentes)
    pessoa = Pessoas(
        cpf=row['CPF'], 
        nome=row['NmeCliente'], 
        conveniada=row['Conveniada'], 
        data_nascimento=row['DtaInscricao'], 
        seqcliente=row['SeqCliente'], 
        tipo_vinculo=row['TipoVinculo'], 
        plano=row['Plano']
    )
    
    if row['TipoVinculo'] == "TITULAR":
        servidor = Servidores(pessoa, cod_matricula=row['CodMatricula'], nroinscricao=row['NroInscricao'])
        titulares.append(servidor)
        servidor_por_matricula[row['CodMatricula']] = servidor  # Mapeia o servidor pela matrícula
    else:
        dependente = Dependentes(pessoa, nroinscricao=row['NroInscricao'])
        
        cod_matricula = row['CodMatricula'] # Adiciona o dependente à lista do servidor correspondente
        if cod_matricula in servidor_por_matricula:
            servidor_por_matricula[cod_matricula].add_dependente(dependente)  # Adiciona o dependente ao servidor
        dependentes.append(dependente)  # Opcional: manter a lista de dependentes, caso precise dela

# Crian um novo DataFrame a partir das classes/listas
dados = []

for servidor in titulares:
    # Adiciona o servidor
    dados.append({
        'Nome': servidor.pessoa.nome,
        'CPF': servidor.pessoa.cpf,
        'Matrícula': servidor.cod_matricula,
        'NroInscricao': servidor.nroinscricao,
        'Vinculo': servidor.pessoa.tipo_vinculo
    })
    
    # Adiciona os dependentes do servidor
    for dependente in servidor.dependentes:
        dados.append({
            'Nome': dependente.pessoa.nome,
            'CPF': dependente.pessoa.cpf,
            'Matrícula': servidor.cod_matricula,
            'NroInscricao': dependente.nroinscricao,
            'Vinculo': dependente.pessoa.tipo_vinculo
        })

# Criação do DataFrame
df_refeito = pd.DataFrame(dados)

# Exibir o DataFrame
display(df_refeito)
```

```python
# Erro na formação da planilha
import pandas as pd

df = pd.read_csv('GEAP.csv', sep=';')
df.columns = df.columns.str.strip()

df_teste = df.head(30)
display(df_teste)

class Pessoas:
    def __init__(self, cpf, nome, conveniada, data_nascimento, seqcliente, tipo_vinculo, plano) -> None:
        self.cpf = cpf
        self.nome = nome
        self.conveniada = conveniada
        self.data_nascimento = data_nascimento
        self.seqcliente = seqcliente
        self.tipo_vinculo = tipo_vinculo
        self.plano = plano
    

class Servidores:
    def __init__(self, pessoa:Pessoas, cod_matricula, nroinscricao) -> None:
        self.pessoa = pessoa
        self.cod_matricula = cod_matricula
        self.nroinscricao = nroinscricao
        self.dependentes = []
        
    def add_dependente(self, dependente):
        # Verificar se o dependente já está vinculado
        if dependente not in self.dependentes:
            self.dependentes.append(dependente)
        else:
            print(f"Dependente {dependente.pessoa.nome} já está vinculado a este titular.")

    def rem_dependente(self, dependente):
        if dependente in self.dependentes:
            self.dependentes.remove(dependente)
        else:
            print(f"Dependente {dependente.pessoa.nome} não está mais vinculado a este titular.")

        
class Dependentes:
    def __init__(self, pessoa:Pessoas,nroinscricao ) -> None:
        self.pessoa = pessoa
        self.nroinscricao = nroinscricao       

# Listas
titulares: list[Servidores] = []
dependentes: list[Dependentes] = []


for i, row in df_teste.iterrows():
    # Criar a pessoa (tanto para titulares quanto dependentes)
    pessoa = Pessoas(
        cpf=row['CPF'], 
        nome=row['NmeCliente'], 
        conveniada=row['Conveniada'], 
        data_nascimento=row['DtaInscricao'], 
        seqcliente=row['SeqCliente'], 
        tipo_vinculo=row['TipoVinculo'], 
        plano=row['Plano']
    )
    
    if row['TipoVinculo'] == "TITULAR":
        # Criar o servidor
        servidor = Servidores(pessoa, cod_matricula=row['CodMatricula'], nroinscricao=row['NroInscricao'])
        titulares.append(servidor)
    else:
        # Criar o dependente
        dependente = Dependentes(pessoa, nroinscricao=row['NroInscricao'])
        dependentes.append(dependente)

# Criar um dicionário para mapear CodMatricula aos titulares/servidores
servidores_por_inscricao = {servidor.nroinscricao: servidor for servidor in titulares}

# Associar os dependentes aos seus respectivos titulares
for dependente in dependentes:
    # Encontra o servidor correspondente
    servidor = servidores_por_inscricao.get(dependente.nroinscricao)
    
    if servidor:
        # Adicionar o dependente ao servidor
        servidor.add_dependente(dependente)
    else:
        print(f"Erro: Não foi encontrado um servidor com a inscrição {dependente.nroinscricaoa} para o dependente {dependente.pessoa.nome}.")

# Cria o DataFrame a partir das classes/listas
dados = []

for servidor in titulares:
    # SeqCliente do titular sempre será 0
    seq_cliente = 0

    # Adiciona o servidor com SeqCliente = 0
    dados.append({
        'Nome': servidor.pessoa.nome,
        'CPF': servidor.pessoa.cpf,
        'Matrícula': servidor.cod_matricula,
        'NroInscricao': servidor.nroinscricao,
        'Vinculo': servidor.pessoa.tipo_vinculo,
        'SeqCliente': seq_cliente,
        'Plano': servidor.pessoa.plano
    })

    # Adiciona os dependentes do servidor com SeqCliente crescente
    for i, dependente in enumerate(servidor.dependentes, start=1):
        dados.append({
            'Nome': dependente.pessoa.nome,
            'CPF': dependente.pessoa.cpf,
            'Matrícula': servidor.cod_matricula,
            'NroInscricao': dependente.nroinscricao,
            'Vinculo': dependente.pessoa.tipo_vinculo,
            'SeqCliente': i ,
            'Plano': dependente.pessoa.plano
        })


# Criação do DataFrame
df_refeito = pd.DataFrame(dados)

# Exibir o DataFrame
display(df_refeito)
```
